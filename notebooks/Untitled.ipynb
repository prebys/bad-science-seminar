```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 153
Ntrial = 1000
sexpect = np.random.binomial(N,.2,Ntrial)
sexpect = sexpect/N
sobs = np.random.binomial(N,.3,Ntrial)
sobs = sobs/N
plt.hist(sexpect,label='What I expect',alpha=.8)
#plt.hist(sobs,label='What I observe',alpha=.8)
plt.xlabel(f'Fraction Correct, {N} tries')
ticks = np.linspace(0.,1.,11)
plt.xticks(ticks)
plt.grid()
plt.legend()
plt.xlim(0,1.)
plt.show()
```
